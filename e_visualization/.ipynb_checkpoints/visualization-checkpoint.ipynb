```python
import pandas as pd

t_df = pd.read_csv("./datasets/titanic.csv")

t_df.head(5)
```

```python
t_df.isna().sum()
```

```python
t_df.Age = t_df.Age.fillna(t_df.Age.mean())
```

```python
t_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 바 플롯 ver.plt - height는 max 값이다
plt.bar(x=t_df.Pclass, height=t_df.Age)
plt.show()

# 바 플롯 ver.sns
sns.barplot(x="Pclass", y="Age", data=t_df)
plt.show()
```

```python
# 바이올린 차트 ver.plt - 나이에 대한 분포도
# plt.violinplot(t_df.Age)
# plt.show()

# 바이올린 차트 ver.sns
# hue로 설정할 feature를 x축으로 설정하면, 여러 개의 군집을 한 번에 볼 수 있다
sns.violinplot(x='Pclass', y="Age", hue='Pclass', data=t_df)
plt.show()
```

```python
# 산점도 ver.plt
plt.scatter(x=t_df.Age, y=t_df.Fare)
plt.show()

# 산점도 ver.sns
sns.scatterplot(x='Age', y='Fare', hue='Pclass', data=t_df)
plt.show()

# 산점도 ver.pandas
t_df.plot(x='Age', y='Fare', kind="scatter")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 히스토그램 ver.sns
# 히스토그램에서 y축은 count 고정이다
# dodge: hue 데이터에 해당하는 막대가 서로 겹치지 않음
sns.histplot(x='Age', data=t_df, hue="Sex", multiple='dodge')
plt.show()

# stack: hue 데이터에 해당하는 막대가 누적됨
sns.histplot(x='Age', data=t_df, hue="Sex", multiple='stack')
plt.show()

# kde: 커널 밀도에 해당하는 선 그래프 표시 (default=False)
sns.histplot(x='Age', data=t_df, kde=True)
plt.show()
```

```python
# 히스토그램 ver.plt
plt.hist(t_df.Age)
plt.show()

# 히스토그램 ver.sns
sns.distplot(x=t_df.Age, kde=True, rug=True)
plt.show()

# 히스토그램 ver.pandas
# grid 옵션을 줄 수 있다
t_df.Age.hist(grid=True)
```

```python
import pandas as pd

a_df = pd.read_csv('./datasets/avocado.csv')

a_df
```

```python
a_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 선 그래프 ver.sns
sns.lineplot(x='Date', y='AveragePrice', data=a_df)
```

```python
# x축에 들어갈 자료가 datetime 형식이 아니면 위와 같은 그래프가 나온다
a_df.loc[:, "Date"] = pd.to_datetime(a_df.Date)

# Date를 datetime 형식으로 바꾸면 아래와 같이 깔끔한 x축 레이블을 얻을 수 있다
line = sns.lineplot(x='Date', y='AveragePrice', data=a_df)
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()
```

```python
# hue로 type을 추가함으로서, type별 선 그래프를 표시 
line = sns.lineplot(x='Date', y='AveragePrice', hue="type", data=a_df)
line.set_xticklabels(line.get_xticklabels(), rotation=45)
plt.show()
```
