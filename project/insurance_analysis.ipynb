### 👓 비교 대상과 특정 그룹간의 상관관계 시각화
- 각 지표의 상위/하위 n%에 속하는 고객만 분류하고,  
  해당 고객들의 나이, 직업, 거주지 등의 분포를 조사했습니다.

#### ❗ 위험 프로필 (Risk Profile)
- 점수가 낮을수록 위험도가 높은 고객이며, 반대로 점수가 높을수록 안전한 성향의 고객임을 의미합니다.
- 지표가 0, 1, 2, 3으로 네 가지 뿐이니, 0점과 3점을 받은 고객만 추려서 성향을 분석했습니다.

#### 🔹 가설
1. 위험 프로필 0점을 받은 고객들 중 음주 운전 또는 난폭 운전 기록이 있는 고객들의 지분이 높을 것이며,  
   이와 반대로, 위험 프로필 3점을 받은 고객들은 사고 기록이 없거나 사고를 당한 쪽이 많을 것으로 추정됩니다.
2. 이혼 또는 실직 등, 현재 부정적인 사건을 겪는 고객들이 위험 프로필 0점에서 높은 지분을 차지하고 있을 것입니다.

```python
# 고객들의 정보 컬럼들
data_columns = [
    'Marital Status', 'Occupation', 'Education', 'City',
    'Policy Type', 'Main Language', 'Driving Record', 'Life Events'
]
```

#### 위험 프로필 0점
- 사고 위험도가 높은 고객들을 의미합니다.

```python
# 위험 프로필 점수가 0점인 고객만 분리
rp_0_df = icp_df[icp_df['Risk Profile'] == 0]

rp_0_df
```

```python
import matplotlib.pyplot as plt

# 플롯 폰트 크기 설정
plt.rc('axes', labelsize=20)

# 서브 플롯 설정
fig, axes = plt.subplots(2, 4, figsize=(20, 12))

# 데이터프레임에서 위의 정보들을 조회하여, 각 정보 별로 파이 차트 생성
for i in range(len(data_columns)):
    target_column = data_columns[i]
    
    # 위험 프로필이 0점인 고객들만 모아놓은 데이터프레임에서 특정 정보들로 groupby 하고, 각 유형별 퍼센티지 산출
    rp_0_count = rp_0_df.groupby(target_column)['Risk Profile'].count()
    rp_0_ratio = rp_0_count / rp_0_count.sum()

    if target_column == "City":
        rp_0_ratio = rp_0_ratio.nlargest(10)

    # 위에서 구한 확률(pandas.Series)을 데이터프레임으로 변환
    pie_df = rp_0_ratio.reset_index()

    # 파이 차트를 그리기 위해 필요한 데이터 가공
    data_labels = pie_df[pie_df.columns[0]]
    data_explode = [0.05 for _ in range(len(data_labels))]
    pie_color = plt.cm.Pastel1(range(len(data_labels)))

    # 차트의 행 지정 - 기본 0, 5번째 차트부터는 1
    ax_row = 0
    ax_col = i

    if i >= 4:
        ax_row = 1
        ax_col = i - 4

    # 특정 위치에 파이 그래프 생성
    axes[ax_row, ax_col].pie(rp_0_ratio, labels=data_labels, autopct='%.1f%%', startangle=90,
            explode=data_explode, shadow=True, colors=pie_color, counterclock=False)

    # 제목 및 레이블 설정
    axes[ax_row, ax_col].set_title(f'Risk Profile 0 Ratio by {target_column}')
    axes[ax_row, ax_col].set_xlabel(target_column)

plt.tight_layout()
plt.show()
```

#### 위험 프로필 3점

```python
# 위험 프로필 점수가 0점인 고객만 분리
rp_3_df = icp_df[icp_df['Risk Profile'] == 3]

rp_3_df
```

```python
import matplotlib.pyplot as plt

# 서브 플롯 설정
fig, axes = plt.subplots(2, 4, figsize=(20, 12))

# 데이터프레임에서 위의 정보들을 조회하여, 각 정보 별로 파이 차트 생성
for i in range(len(data_columns)):
    target_column = data_columns[i]
    
    # 위험 프로필이 3점인 고객들만 모아놓은 데이터프레임에서 특정 정보들로 groupby 하고, 각 유형별 퍼센티지 산출
    rp_3_count = rp_3_df.groupby(target_column)['Risk Profile'].count()
    rp_3_ratio = rp_3_count / rp_3_count.sum()

    # 도시 컬럼은 정보가 많으니 상위 10개만 분리
    if target_column == "City":
        rp_3_ratio = rp_3_ratio.nlargest(10)

    # 위에서 구한 확률(pandas.Series)을 데이터프레임으로 변환
    pie_df = rp_3_ratio.reset_index()

    # 파이 차트를 그리기 위해 필요한 데이터 가공
    data_labels = pie_df[pie_df.columns[0]]
    data_explode = [0.05 for _ in range(len(data_labels))]
    pie_color = plt.cm.Pastel1(range(len(data_labels)))

    # 차트의 행 지정 - 기본 0, 5번째 차트부터는 1
    ax_row = 0
    ax_col = i

    if i >= 4:
        ax_row = 1
        ax_col = i - 4

    # 특정 위치에 파이 그래프 생성
    axes[ax_row, ax_col].pie(rp_3_ratio, labels=data_labels, autopct='%.1f%%', startangle=90,
            explode=data_explode, shadow=True, colors=pie_color, counterclock=False)

    # 제목 및 레이블 설정
    axes[ax_row, ax_col].set_title(f'Risk Profile 3 Ratio by {target_column}')
    axes[ax_row, ax_col].set_xlabel(target_column)

plt.tight_layout()
plt.show()
```

#### 💰 1회 보상 한도 (Coverage Amount)
- 상위 / 하위 10%에 속하는 고객들의 특성을 분석했습니다.

#### 🔹 가설
1. 사고 기록이 없는 고객들이 1회 보상 한도가 높을 것으로 예상됩니다.
2. 또한, 가족 단위로 가입한 고객들이 보상 한도가 높게 나타날 것입니다.

#### 1회 보상 한도 상위 10%

```python
# 데이터프레임에서 Coverage Amount 수치 상위 10% 값 출력
ca_top_10_threshold = icp_df['Coverage Amount'].quantile(0.9)

ca_top_10_threshold
```

```python
# 1회 보상 한도 상위 10%에 해당하는 고객들만 분리
ca_top_10_df = icp_df[icp_df['Coverage Amount'] >= ca_top_10_threshold]

ca_top_10_df
```

```python
import matplotlib.pyplot as plt

# 서브 플롯 설정
fig, axes = plt.subplots(2, 4, figsize=(20, 12))

# 데이터프레임에서 위의 정보들을 조회하여, 각 정보 별로 파이 차트 생성
for i in range(len(data_columns)):
    target_column = data_columns[i]
    
    # 1회 보상 한도 상위 10%인 고객들만 모아놓은 데이터프레임에서 특정 정보들로 groupby 하고, 각 유형별 퍼센티지 산출
    ca_top_10_count = ca_top_10_df.groupby(target_column)['Coverage Amount'].count()
    ca_top_10_ratio = ca_top_10_count / ca_top_10_count.sum()

    # 도시 컬럼은 정보가 많으니 상위 10개만 분리
    if target_column == "City":
        ca_top_10_ratio = ca_top_10_ratio.nlargest(10)

    # 위에서 구한 확률(pandas.Series)을 데이터프레임으로 변환
    pie_df = ca_top_10_ratio.reset_index()

    # 파이 차트를 그리기 위해 필요한 데이터 가공
    data_labels = pie_df[pie_df.columns[0]]
    data_explode = [0.05 for _ in range(len(data_labels))]
    pie_color = plt.cm.Pastel1(range(len(data_labels)))

    # 차트의 행 지정 - 기본 0, 5번째 차트부터는 1
    ax_row = 0
    ax_col = i

    if i >= 4:
        ax_row = 1
        ax_col = i - 4

    # 특정 위치에 파이 그래프 생성
    axes[ax_row, ax_col].pie(ca_top_10_ratio, labels=data_labels, autopct='%.1f%%', startangle=90,
            explode=data_explode, shadow=True, colors=pie_color, counterclock=False)

    # 제목 및 레이블 설정
    axes[ax_row, ax_col].set_title(f'Top 10% of Coverage Amount Ratio by {target_column}')
    axes[ax_row, ax_col].set_xlabel(target_column)

plt.tight_layout()
plt.show()
```
