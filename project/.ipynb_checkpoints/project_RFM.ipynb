```python
import pandas as pd

pd.set_option('display.max_columns', None)

# Ï†ÑÏ≤¥ ÌåêÎß§ Îç∞Ïù¥ÌÑ∞Í∞Ä Îã¥Í∏¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î∂àÎü¨Ïò§Í∏∞
sd_df = pd.read_csv('./datasets/rfm/Sales_domain.csv', encoding='cp1252')

sd_df
```

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï†ïÎ≥¥ Ï∂úÎ†•
sd_df.info()
```

### Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨

#### Ïª¨Îüº Ïû¨Î∞∞Ïπò

```python
# ÌäπÏ†ï Ïú†ÌòïÎ≥ÑÎ°ú Ïª¨Îüº Ïû¨Î∞∞Ïπò
updated_index = [
    'Date', 'product_code', 'division', 'segment', 'product', 'category',
    'variant', 'manufacturing_cost', 'gross_price', 'cost_year', 'customer_code',
    'customer', 'channel', 'platform', 'region', 'sub_zone',
    'market', 'sold_quantity', 'pre_invoice_discount_pct', 'fiscal_year'
]

sd_df = sd_df.reindex(columns=updated_index)

sd_df
```

#### Í≤∞Ï∏°Ïπò Í≤ÄÏÇ¨

```python
# Ïª¨Îüº Î≥Ñ Í≤∞Ï∏°Ïπò Ïàò Ï∏°Ï†ï
sd_df.isna().sum()
```

```python
# Í≤∞Ï∏°Ïπò Î≥¥Ïú† Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
sd_df[sd_df['region'].isna()]
```

```python
# Í≤∞Ï∏°Ïπò Î≥¥Ïú† Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú
sd_df = sd_df[~ sd_df['region'].isna()]

sd_df
```

```python
# ÎÇ®ÏùÄ Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäîÏßÄ ÌïôÏù∏
sd_df.isna().sum()
```

#### Ï§ëÎ≥µÌñâ Í≤ÄÏÇ¨

```python
# Ï§ëÎ≥µÎêòÎäî Îç∞Ïù¥ÌÑ∞Í∞Ä ÏûàÎäîÏßÄ ÌôïÏù∏
sd_df.duplicated().sum()
```

#### Ïù¥ÏÉÅÏπò(Outliner) Í≤ÄÏÇ¨

```python
# Ïà´ÏûêÌòï Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏßÑ Ïª¨ÎüºÏùò ÏàòÏπò Ï†ïÎ≥¥Îì§ Ï∂úÎ†•
sd_df.describe().T
```

```python
# ÎÇ©Ìíà ÏàòÎüâÏù¥ 0Ïù∏ Îç∞Ïù¥ÌÑ∞Îì§ ÏÇ≠Ï†ú
sd_df = sd_df[~ (sd_df['sold_quantity'] == 0)]

sd_df
```

```python
# ÎÇ©Ìíà ÏàòÎüâ Ïª¨ÎüºÏùò ÏàòÏπò Ï†ïÎ≥¥Îì§ Ïû¨Ï∂úÎ†•
sd_df[['sold_quantity']].describe().T
```

```python
# Î∂àÌïÑÏöîÌïú Ïª¨Îüº Ï†úÍ±∞
drop_columns = ['division', 'segment', 'sub_zone'] 

sd_df = sd_df.drop(labels=drop_columns, axis=1)

sd_df
```

```python
# ÎÇ©Ìíà ÏùºÏûê(Date) Ïª¨ÎüºÏùÑ datetime ÌòïÏãùÏúºÎ°ú Î≥ÄÍ≤Ω
sd_df['Date'] = pd.to_datetime(sd_df['Date'], dayfirst=True)

sd_df
```

```python
(sd_df[sd_df['customer'] == 'Amazon']).count()
```

```python
(sd_df[sd_df['customer'] == 'Amazon ']).count()
```

#### ‚ö†Ô∏è Ïù¥Ïäà Î∞úÏÉù
- ÌòÑÏû¨ customer Ïª¨ÎüºÏùò Í∞íÏù¥ 'Amazon' Ïù∏ Í≤ÉÍ≥º 'Amazon ' Ïù∏ Í≤ÉÏù¥ ÎèôÏãúÏóê Ï°¥Ïû¨ÌïòÎäî Í≤ÉÏùÑ ÌôïÏù∏
- Îî∞ÎùºÏÑú 'Amazon ' Í∞íÏùÑ 'Amazon' ÏúºÎ°ú ÏàòÏ†ïÌï®ÏúºÎ°úÏÑú Ïª¨Îüº ÎÇ¥Ïö©Ïóê Î∞úÏÉùÌïú Ïù¥ÏÉÅÏùÑ Ìï¥Í≤∞

```python
# 'Amazon ' Í∞íÏùÑ 'Amazon'ÏúºÎ°ú Î≥ÄÍ≤Ω ÌõÑ Ï†ÅÏö© Ïó¨Î∂Ä ÌôïÏù∏
sd_df['customer'] = sd_df['customer'].replace('Amazon ', 'Amazon')

sd_df['customer'].value_counts()
```

```python
# Ïù∏Îç±Ïä§ Ï¥àÍ∏∞Ìôî
sd_df = sd_df.reset_index(drop=True)

sd_df
```

```python
# Í∞Å Îç∞Ïù¥ÌÑ∞ Î≥Ñ Ï¥ù Í∞ÄÍ≤© Íµ¨ÌïòÍ∏∞
# Ï†úÌíà Î≥Ñ Ï†úÏ°∞ÎπÑ * ÎÇ©Ìíà ÏàòÎüâ * (1 - Ìï†Ïù∏Ïúª)
sd_df.loc[:, 'total_price'] = sd_df['manufacturing_cost'] * sd_df['sold_quantity'] \
                                * (1 - sd_df['pre_invoice_discount_pct'])

sd_df
```

```python
# Ï¥ù Í∞ÄÍ≤©ÏùÑ ÏÜåÏàòÏ†ê 3ÏûêÎ¶¨ÍπåÏßÄÎßå ÌëúÏãú
sd_df['total_price'] = sd_df['total_price'].round(3)

sd_df
```

### RFM Ï†êÏàò ÏÇ∞Ï∂ú
- Í∞Å Ï†êÌè¨ Î≥Ñ RFM Ï†êÏàò ÏÇ∞Ï∂ú

```python
# Date ÏµúÏã†ÏàúÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ïû¨Ï†ïÎ†¨ - ÎßàÏßÄÎßâ Í±∞ÎûòÏùºÏùÄ 2021-12-01 
sd_df.sort_values(['Date'], ascending=False)
```

```python
# Recency Ïª¨ÎüºÏùÑ Ï∂îÍ∞ÄÌïòÍ∏∞ Ï†ÑÏóê, Í∏∞Ï°¥ Í≤ÉÏùÑ Î≥µÏÇ¨Ìïú ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
r_sd_df = sd_df.copy()

r_sd_df
```

```python
# 2022-01-01ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÎßàÏßÄÎßâ Í±∞ÎûòÏùºÎ°úÎ∂ÄÌÑ∞ ÏßÄÎÇú ÏãúÍ∞ÑÏùÑ Recency Ïª¨ÎüºÏóê Ï∂îÍ∞Ä
r_sd_df.loc[:, 'Recency'] = r_sd_df['Date'].apply(lambda x: (pd.to_datetime('2022-01-01') - x).days)

r_sd_df
```

```python
# Recency Ïª¨Îüº Ï∂îÍ∞Ä Î∞è Í∞í Ï†ÅÏö© Ïó¨Î∂Ä ÌôïÏù∏
r_sd_df.sort_values(['Recency'])
```

```python
# Date Ïª¨ÎüºÏùÄ Îçî Ïù¥ÏÉÅ ÌïÑÏöî ÏóÜÏúºÎãà ÏÇ≠Ï†ú
r_sd_df = r_sd_df.drop(['Date'], axis=1)

r_sd_df
```

```python
# ÏúÑÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Î∞îÌÉïÏúºÎ°ú Í∞Å Ï†êÌè¨(customer) Î≥Ñ RFM Ï†êÏàò ÏÇ∞Ï∂ú
rfm_sd_df = r_sd_df.groupby('customer')\
                .agg(Recency=('Recency', 'min'), Frequency=('customer', 'count'), Monetary=('total_price', 'sum'))

rfm_sd_df
```

#### Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî
- Recency, Frequency, Monetary Í∞íÏùÑ Ï†êÎ∂Ä 0~1 ÏÇ¨Ïù¥Ïùò Í∞íÏúºÎ°ú Î≥ÄÌôò

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScalerÎ•º ÌÜµÌïú RFM Îç∞Ïù¥ÌÑ∞Ïùò Ïä§ÏºÄÏùºÎßÅ Ï§ÄÎπÑ
scaler = MinMaxScaler()
```

```python
# Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî Ï†Ñ, Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î≥µÏÇ¨
scaled_rfm_df = rfm_sd_df.copy()

scaled_rfm_df
```

```python
# RFM Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî
scaled_rfm_df[['Recency', 'Frequency', 'Monetary']] = \
                            scaler.fit_transform(scaled_rfm_df[['Recency', 'Frequency', 'Monetary']])

scaled_rfm_df
```

```python
# Recency Í∞í Î≥ÄÍ≤Ω
scaled_rfm_df['Recency'] = 1 - scaled_rfm_df['Recency']

scaled_rfm_df
```

```python
# RFM Ï†êÏàò Ìï©ÏÇ∞ Í≤∞Í≥ºÎ•º ÏÉà Ïª¨ÎüºÏóê Ï∂îÍ∞Ä
scaled_rfm_df.loc[:, 'total_score'] = scaled_rfm_df['Recency'] \
                                    + scaled_rfm_df['Frequency'] \
                                    + scaled_rfm_df['Monetary']

scaled_rfm_df
```

```python
# Îç∞Ïù¥ÌÑ∞ Î≥ëÌï© Ï†Ñ, customer Ïª¨ÎüºÏùÑ Ïù∏Îç±Ïä§Î°ú Î∂ÑÎ¶¨
scaled_rfm_df = scaled_rfm_df.reset_index()

scaled_rfm_df
```

```python
# Í∏∞Ï°¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Î≥ëÌï©
sd_rfm_df = sd_df.merge(scaled_rfm_df[['customer', 'total_score']], on='customer')

sd_rfm_df
```

#### Ï†êÌè¨ Î≥Ñ Îì±Í∏â Íµ¨Î∂Ñ
- total_score Í∞íÏóê Îî∞Îùº Ï†êÌè¨ Î≥Ñ Îì±Í∏âÏùÑ Íµ¨Î∂Ñ

```python
import numpy as np

# Í∞Å Ï†êÌè¨Ïùò RFM Ï¥ùÏ†êÏùÑ ÏùºÏ†ï ÌçºÏÑºÌã∞ÏßÄÎ°ú Íµ¨Î∂Ñ
l1, l2, l3, l4 = np.percentile(sd_rfm_df['total_score'], [20, 35, 65, 80])

print(l1, l2, l3, l4)
```

```python
# ÏúÑÏùò ÌçºÏÑºÌã∞ÏßÄ Íµ¨Í∞ÑÏóê Îî∞Îùº ÌöåÏõê Îì±Í∏âÏùÑ ÎÇòÎà†Ï£ºÎäî Ìï®Ïàò ÏÑ†Ïñ∏
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
# Í∞Å Ï†êÌè¨Ïùò RFM Ï¥ùÏ†êÏóê ÏúÑÏùò Ìï®ÏàòÎ•º Ï†ÅÏö©Ìïú Ï†êÌè¨ Îì±Í∏â Ïª¨Îüº ÏÉùÏÑ±
sd_rfm_df['level'] = sd_rfm_df['total_score'].apply(get_level)

sd_rfm_df
```

```python
# ÌòÑÏû¨ Ï†ïÏàòÌòïÏúºÎ°ú ÎêòÏñ¥ÏûàÎäî ÌöåÏõê Îì±Í∏âÏùÑ Î¨∏ÏûêÏó¥Î°ú ÏπòÌôòÌïòÏó¨ Í∞ÄÎèÖÏÑ± Ìñ•ÏÉÅ
sd_rfm_df['level'] = sd_rfm_df['level'].replace([5, 4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])

sd_rfm_df
```

```python
# ÌïÑÏöîÏóÜÎäî Ïª¨Îüº ÏÇ≠Ï†ú
sd_rfm_df = sd_rfm_df.drop(['total_score'], axis=1)

sd_rfm_df
```

```python
# market Ïª¨ÎüºÏùò Ïª¨ÎüºÎ™ÖÏùÑ countryÎ°ú Î≥ÄÍ≤Ω
sd_rfm_df = sd_rfm_df.rename(columns={'market': 'country'})

sd_rfm_df
```

```python
# region Ïª¨ÎüºÏóê Îì§Ïñ¥Í∞ÄÎäî Í∞íÏùÑ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥ÄÍ≤Ω
sd_rfm_df['region'] = sd_rfm_df['region']\
                        .replace(['APAC', 'EU', 'LATAM'], ['Asia & Pacific', 'Europe', 'Latin America'])

sd_rfm_df
```
