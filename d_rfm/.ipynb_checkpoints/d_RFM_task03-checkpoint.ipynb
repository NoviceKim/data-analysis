### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석

> invoice_no: 송장 번호. 명사 같은. 각 작업에 고유하게 할당된 문자 'I'와 6자리 정수의 조합  
customer_id: 고객번호. 명사 같은. 각 작업에 고유하게 할당된 문자 'C'와 6자리 정수의 조합  
gender: 고객 성별의 문자열 변수  
age: 고객 연령의 양의 정수 변수  
category: 구매한 상품 카테고리의 문자열 변수 
quantity: 거래당 각 제품(항목)의 수량  
price: 단가. 숫자. 단위당 제품 가격은 터키 리라(TL)  
payment_method: 거래에 사용된 결제 방법(현금, 신용 카드 또는 직불 카드)의 문자열 변수  
invoice_date: 송장 날짜. 거래가 발생한 날  
shopping_mall: 거래가 이루어진 쇼핑몰 이름의 문자열 변수  

### 데이터프레임 전처리

#### 데이터프레임 읽기

```python
import pandas as pd

# 데이터프레임 읽기 - 여러 쇼핑몰의 거래 내역을 모아놓은 데이터
cs_df = pd.read_csv('./datasets/customer_shopping_data.csv')

cs_df
```

```python
# 데이터프레임 정보 출력
cs_df.info()
```

#### 결손 / 중복 데이터 확인

```python
# 각 컬럼 별 결손 데이터 수 확인 - 결손 데이터 없음
cs_df.isna().sum()
```

```python
# 중복값 확인 - 중복값 없음
cs_df.duplicated().sum()
```

#### 이상치 관리

```python
# 각 컬럼의 수치 데이터 출력
cs_df.describe().T
```

```python
# 가격 컬럼에서 최소값(5.23리라, 한화 약 200원)을 가진 행 출력 
cs_df[cs_df['price'] == cs_df['price'].min()]
```

#### 중간 보고

- 이 데이터프레임에는 결손 또는 중복되는 데이터가 없는 것으로 확인되었습니다.
<br/><br/>
- 또한, 이상치를 발견하고 관리하기 위해 수치 데이터를 출력해본 결과에 의하면  
  사용자들의 나이 및 구매 수량 역시 별다른 이상치가 없는 것으로 판단됩니다.
- 가격 컬럼의 최소값(min)은 5.23 리라, 한화 약 200원으로 측정되었기 때문에 해당 값을 가진 행들을 조사해 본 결과  
  해당 데이터들은 식음료를 구매한 내역이며, 현금으로 거래된 내역이 다수 존재하는 것이 확인되었습니다.
- 이러한 거래는 현실에서도 충분히 있을법하다고 판단, 따라서 해당 데이터들 역시 이상치로 간주하지 않으며  
  이 데이터프레임에는 이상치가 없는 것으로 결론 내렸습니다.

#### 각 거래 건의 총 구매액 계산

```python
# 각 거래 건의 구매 수량 * 단가의 값인 총 구매액을 'total_price'라는 새로운 컬럼에 추가
cs_df.loc[:, "total_price"] = cs_df['quantity'] * cs_df['price']

cs_df
```

#### 날짜 데이터 형식 수정

```python
# 날짜 데이터를 추가하기 전, 원본 데이터프레임 복사
cs_days_df = cs_df.copy()

cs_days_df
```

```python
# 원본 데이터프레임의 날짜 컬럼을 datetime 형식으로 변경하여, 복사된 데이터프레임의 날짜 컬럼에 할당
# 이 때, dayfirst=True로 하여 원래 컬럼에 있던 값을 일/월/연도 순으로 취급, 이것을 연도-월-일 순으로 바꿔준다
cs_days_df['invoice_date'] = pd.to_datetime(cs_df['invoice_date'], dayfirst=True)

cs_days_df
```

```python
# 날짜 컬럼의 현재 데이터 타입 확인 - 현재 날짜 타입
cs_days_df['invoice_date'].dtype
```

```python
# 날짜를 내림차순(최신순)으로 정렬 - 최근 거래 일자는 2023-03-08
cs_days_df.sort_values(['invoice_date'], ascending=False)
```

```python
# 최신 거래일로부터 하루 지난 2023-03-09를 기준으로 거래일로부터 지난 날짜 수를 계산하여
# 'date_passed' 라는 이름의 새로운 컬럼을 생성 후 내용 추가
cs_days_df.loc[:, 'date_passed'] = cs_days_df['invoice_date']\
                                    .apply(lambda x: (pd.to_datetime('2023-03-09') - x).days)

cs_days_df
```

```python
# invoice_date 컬럼은 더 이상 필요 없으니 삭제
cs_days_df = cs_days_df.drop(['invoice_date'], axis=1)

cs_days_df
```

#### RFM 점수 계산

```python
# 각 회원 별 rfm 점수만 따로 분리한 새로운 데이터프레임 생성
rfm_df = cs_days_df.groupby('customer_id')\
            .agg(Recency=('date_passed', 'min'), Frequency=('customer_id', 'count'), Monetary=('total_price', 'sum'))

rfm_df
```

```python
# Frequency의 값 분포 출력 - 모든 행의 Frequency 값이 1인 것을 확인
rfm_df['Frequency'].value_counts()
```

```python
# cs_days_df에서 각 회원의 value_count 출력 - 모든 회원이 단 한 번의 거래만 한 것을 알 수 있다
cs_days_df['customer_id'].value_counts()
```

#### Frequency 값 재정의
- 각 회원 별 거래 횟수로는 Frequency 지표를 구분할 수 없으니, 한 번의 거래 시 구매한 물품 수로 구분

```python
# RFM 데이터프레임과 기존 데이터프레임을 병합한 새로운 데이터프레임 생성
# 휴먼 에러로 인한 원치 않는 데이터 유실을 방지하기 위함 
merged_df = rfm_df.merge(cs_days_df, on='customer_id')

merged_df
```

```python
# quantity 값을 Frequency에 그대로 할당
merged_df['Frequency'] = merged_df['quantity']

merged_df
```

```python
# Frequency 값이 수정된 새로운 RFM 데이터프레임 생성
new_rfm_df = merged_df[['customer_id', 'Recency', 'Frequency', 'Monetary']]

new_rfm_df
```

#### 데이터 정규화

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler를 통한 RFM 데이터의 스케일링 준비
normailzer = MinMaxScaler()
```

```python
# 기존 데이터프레임의 R, F, M 컬럼에 정규화 실행
new_rfm_df[['Recency', 'Frequency', 'Monetary']] = \
                                        normailzer.fit_transform(new_rfm_df[['Recency', 'Frequency', 'Monetary']])

new_rfm_df
```

```python
# 최근에 거래한 사람일수록 Recency 점수가 높아야 하므로, 1에서 기존 Recency 값을 빼준다
new_rfm_df['Recency'] = 1 - new_rfm_df['Recency']
```

```python
# RFM 점수를 합산한 결과를 새 컬럼에 추가
new_rfm_df.loc[:, 'total_score'] = new_rfm_df['Recency'] + new_rfm_df['Frequency'] + new_rfm_df['Monetary']

new_rfm_df
```

```python
# 원본 데이터프레임에 RFM 테이블 병합
cs_rfm_df = cs_df.merge(new_rfm_df[['customer_id', 'total_score']], on='customer_id')

cs_rfm_df
```

#### 회원 등급 구분

```python
import numpy as np

# 각 회원의 RFM 총점을 일정 퍼센티지로 구분
l1, l2, l3, l4 = np.percentile(cs_rfm_df['total_score'], [10, 35, 65, 90])

print(l1, l2, l3, l4)
```

```python
# 위의 퍼센티지 구간에 따라 회원 등급을 나눠주는 함수 선언
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
# 각 회원의 RFM 총점에 위의 함수를 적용한 회원 등급 컬럼 생성
cs_rfm_df['level'] = cs_rfm_df['total_score'].apply(get_level)

cs_rfm_df
```

```python
# 현재 정수형으로 되어있는 회원 등급을 문자열로 치환하여 가독성 향상
cs_rfm_df['level'] = cs_rfm_df['level'].replace([5, 4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])

cs_rfm_df
```

```python
# total_score 컬럼은 더 이상 필요 없으니 삭제
cs_rfm_df = cs_rfm_df.drop(['total_score'], axis=1)

cs_rfm_df
```

#### 중간보고

- 각 구매 건 별로 총 구매액과 2023-03-09 기준 마지막 거래일로부터 지난 시간(일)을 구하여  
  새로운 컬럼에 해당 데이터들을 추가했습니다.
<br/><br/>
- 이를 바탕으로 각 회원 별 RFM 점수를 산출하려 했으나 모든 회원의 Frequency 값이 1로 동일한 현상을 확인했습니다.
- 원인을 조사해보니, 이는 모든 회원이 단 한 번의 거래만을 했기 때문이었으며  
  이 시점에서 Frequency를 산정하는 기준을 바꿔야겠다고 판단했습니다.
<br/><br/>
- 따라서 각 회원 별 빈도를 구분할 수 있는 값인 quantity(구매 수량) 컬럼을 새로운 Frequency 선정 기준으로 설정했습니다.
- 위 과정에 따라 구매 수를 새로운 Frequency로 설정한 결과, 각 회원의 RFM 총점이 서로 유의미한 구분을 보이는 것을 확인했습니다.

### 데이터 시각화

#### 각 등급 별 회원 분포

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 각 등급 별 회원 분포를 countplot으로 시각화

order = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(x='level', data=cs_rfm_df, palette='muted', order=order)
plt.title('Customers by Level')
plt.show()
```

#### 데이터 분석 내용
- 각 카테고리 별 거래 건수와 총 수익 비교 
  > 총 수익이 가장 많은 상위 3개 카테고리에 소비한 회원들의  
  > 성비, 연령대, 등급, 주 거래처(쇼핑몰)을 분석하여 마케팅 방향 제시

```python
# 각 카테고리 별 거래 건수 출력
cs_rfm_df['category'].value_counts()
```

```python
# 각 카테고리 별 총 매출액 출력
cs_rfm_df.groupby('category')['total_price'].sum()
```

```python
# 위 2가지 지표를 데이터프레임으로 만든 다음 병합

# 지수표기법이 아닌, 원래 숫자를 표시하기 위해 pandas의 float값 출력 형식을 변경 
pd.options.display.float_format = '{:.2f}'.format

category_count_df = cs_rfm_df['category'].value_counts().reset_index()
category_total_price_df = cs_rfm_df.groupby('category')['total_price'].sum().reset_index()

category_df = category_count_df.merge(category_total_price_df, on="category")

category_df = category_df.rename(columns={"category": "Category", "count": "Count", "total_price": "Take"})

category_df
```

```python
# 위 데이터프레임의 통계값을 barplot으로 시각화
fig, axes = plt.subplots(1, 2, figsize=(20, 12))

sns.set(font_scale=1.8)

for i, col in enumerate(['Count', 'Take']):
    sns.barplot(ax=axes[i], x="Category", y=col, data=category_df, palette='muted')
    axes[i].set_title(f'Total {col} for Each Categories')
    axes[i].set_xlabel(col)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

#### 중간 보고
- 본격적인 분석에 앞서, 모든 쇼핑몰의 카테고리 별 거래 건수와 총 매출액을 시각화 했습니다.
<br/><br/>
- 그 결과 의류 품목이 가장 많이 거래되었다는 걸 알 수 있었고,  
  그에 따른 매출액 역시 가장 높았다는 사실을 확인할 수 있었습니다.
<br/><br/>
- 두 번째로 매출이 높았던 것은 신발 품목이었으며, 거래 건수 순위는 높지 않았던 것으로 보아  
  건 당 거래액이 높았던 것으로 추정됩니다.
  따라서 이 제품군을 구매한 사람들의 성별, 연령대 및 회원 등급에 대해 자세한 조사가 필요할 것으로 보입니다.
<br/><br/>
- 세 번째로 매출이 높았던 것은 가전 품목으로, 마찬가지로 거래 건수는 많지 않기 때문에
