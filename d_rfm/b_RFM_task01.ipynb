#### Numpy를 이용한 결제 금액의 이상치 조사

```python
import numpy as np

# 결제 금액의 평균, 표준편차, 최소/최대값, 사분위수 출력
co_df[['TransactionAmount']].describe().T
```

```python
# 사분위수를 이용하여 정상 범위 출력
Q1 = co_df.describe().T.loc['TransactionAmount', '25%']
Q3 = co_df.describe().T.loc['TransactionAmount', '75%']

iqr = Q3 - Q1

print(f'1분위수: {Q1} / 3분위수: {Q3} / 사분위수 범위: {iqr}')

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

# 결제 금액이 0보다 작을 수는 없으므로, 0보다 작은 데이터를 0으로 취급
if lower_bound < 0:
    lower_bound = 0

print(f'정상 범위: {lower_bound} ~ {upper_bound}')
```

```python
# 이상치 출력 - 이상치 없음
co_df[(co_df['TransactionAmount'] < lower_bound) | (co_df['TransactionAmount'] > upper_bound)]
```

### 현재 구현 가능한 분석 방향
- 특정 제품이 어디에서 많이 팔렸는지(제품 별 총 결제 금액)를 조사하여 지역 별 마케팅 방향 잡기
- 지역 별 총 결제 금액을 조사히고, 각 지역 별 구매한 제품의 비율(파이 그래프)을 분석하여, 지역 별로 판매가 부진한 상품들의 마케팅 촉진

#### R (Recency)
- 얼마나 최근에 구매했는지를 나타내는 지표
- 기존의 날짜 데이터를 정수 값으로 변환
- 오랫동안 구매하지 않았을 때 마케팅 집중하기

```python
from datetime import datetime

# 오늘 날짜
today = pd.to_datetime(datetime.today().date())
```

```python
# 구매 날짜 컬럼의 dtype을 datetime 형식으로 변환
co_df['PurchaseDate'] = pd.to_datetime(co_df['PurchaseDate'])
```

```python
# 오늘 날짜와 구매한 날짜의 차이를 Recency 컬럼으로 저장
co_df['Recency'] = (today - co_df['PurchaseDate']).dt.days

co_df
```

#### F (Frequency)
- 얼마나 자주 구매했는지를 나타내는 지표
- 각 상품 별 구매 횟수를 F 지표로서 사용

```python
# 각 상품 별 구매 횟수
co_df['ProductInformation'].value_counts()
```

```python
# 각 상품의 종류에 따라 다른 Frequency를 추가하는 함수
def get_frequency(x):
    if x == "Product A":
        return 225
    if x == "Product B":
        return 245
    if x == "Product C":
        return 277
    if x == "Product D":
        return 253
```

```python
# Frequency 컬럼 추가 후 DataFrame 확인
co_df['Frequency'] = co_df['ProductInformation'].apply(get_frequency)

co_df
```

```python
# 결손 여부 출력
co_df.isnull().sum()
```

#### M (Monetary)
- 얼마나 많은 금액을 투자했는지 나타내는 지표
- 결제 금액(TransactionAmount) 컬럼이 있기 때문에, 해당 컬럼의 데이터를 그대로 사용

```python
co_df['Monetary'] = co_df['TransactionAmount']

co_df
```

#### RFM 점수 산출
- Recency, Frequency, Monetary 컬럼만 별도의 DataFrame으로 따로 분리
- 그 후 정규화를 통해 데이터의 값을 평등하게 맞춰준 후, 해당 값들을 전부 합산하여 RFM 점수 산출

```python
# RFM 컬럼 분리
co_rfm_df = co_df[['Recency', 'Frequency', 'Monetary']]

co_rfm_df
```

#### 정규화

```python
from sklearn.preprocessing import MinMaxScaler

# 모든 데이터를 0~1 사이의 값으로 맞춰주기 위한 MinMaxScaler
normalization = MinMaxScaler()

# 데이터 정규화 준비
rfm_normalization = normalization.fit_transform(co_rfm_df)
```

```python
rfm_normalization
```

```python
# 기존 RFM DataFrame의 컬럼의 값을 정규화된 값으로 변환 
co_rfm_df = pd.DataFrame(rfm_normalization, columns=co_rfm_df.columns)

co_rfm_df
```

```python
# RFM 점수 컬럼을 원래 DataFrame에 추가
co_df['TotalScore'] = co_rfm_df['Recency'] + co_rfm_df['Frequency'] + co_rfm_df['Monetary']

co_df
```

```python
# 원래 DataFrame에 있는 RFM 컬럼 제거
co_df = co_df.drop(columns=['Recency', 'Frequency', 'Monetary'], axis=1)

co_df
```

```python
import numpy as np

# 점수에 따라 서로 다른 레벨을 부여하기 전에, 각 퍼센티지의 기준값 구하기
l1, l2, l3 = np.percentile(co_df['TotalScore'], [15, 50, 85])

print(l1, l2, l3)
```

```python
# 점수에 따라 서로 다른 레벨을 부여하는 함수
def get_level(x):
    if x <= l1:
        return 4
    if x > l1 and x <= l2:
        return 3
    if x > l2 and x <= l3:
        return 2
    return 1
```

```python
# 점수에 함수 적용 후, 결과값을 Level 컬럼에 추가
co_df['Level'] = co_df['TotalScore'].apply(get_level)

co_df
```

```python
# Level 컬럼의 숫자 값에 따라 서로 다른 문자열을 할당하여 확실한 의미 구분 
co_df['Level'] = co_df['Level'].replace([4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Platinum'])

co_df
```

```python
# 각 등급 별 회원 수
print(co_df['Level'].value_counts())
```
